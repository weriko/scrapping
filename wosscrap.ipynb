```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from contextlib import contextmanager
from selenium.webdriver.chrome.options import Options
import os
import logging

logging.basicConfig(filename='wos.log', filemode='w+', format='%(name)s - %(levelname)s - %(message)s')


import time
import getpass 


#It doesn't seem like WOSs system is smart enough to detect basic automation

def loginUser():
  
    someVariable = getpass.getpass("Press Enter after You are done logging in")

class WOSDownloader:
    def __init__(self, driver = None):
        if driver is None:
            chromedriver = "chromedriver"
            os.environ["webdriver.chrome.driver"] = chromedriver
            chrome_options = Options()

  
            prefs = {'profile.default_content_setting_values.automatic_downloads': 1}
            chrome_options.add_experimental_option("prefs", prefs)
            driver = webdriver.Chrome(chrome_options=chrome_options)
        self.driver = driver
        
    def step_executor(self, steps, tries=3):
        current = 0
        while current<len(steps):
            flag = False
            t = 0
            while t<tries:
                try:
                    steps[current]()
                    flag = True
                    current+=1
                    t=tries
                except Exception as e:
                    print(e)
                    
            if not flag:
                break
                return
        return
        

    def download_file(self, currentfrom, currentto, tries=3):
        
        def step_1():
            exportbutton_text = '//*[@id="snRecListTop"]/app-export-menu/div/button'
            exportbutton = self.driver.find_element("xpath", exportbutton_text )
            exportbutton.click()
            time.sleep(1)
            
        def step_2():
            excelbutton_text = "exportToExcelButton"
            excelbutton = self.driver.find_element("id", excelbutton_text)
            excelbutton.click()
            time.sleep(1)

        def step_3():
            recordsfrom_text = '//*[@id="radio3"]/label/span[2]'
            recordsfrom = self.driver.find_element("xpath", recordsfrom_text)
            recordsfrom.click()
            time.sleep(1)
            
        def step_4():
            firstinput_text = "markFrom"
            firstinput=self.driver.find_element("name", firstinput_text)
         
            self.driver.execute_script(f"arguments[0].value = '{currentfrom}'", firstinput)
            time.sleep(1)

        def step_5():
            secondinput_text = "markTo"
            secondinput= self.driver.find_element("name", secondinput_text)
         
            self.driver.execute_script(f"arguments[0].value = '{currentto}'", secondinput)
            time.sleep(1)
        def step_6():
            recordcontent_text = "Author, Title, Source"
            recordcontent = self.driver.find_element("xpath", "//button[.='" + recordcontent_text + "']")
            recordcontent.click()
            time.sleep(1)
            
        def step_7():
            full_text = "Full Record"
            full = self.driver.find_element("xpath", "//div[.='" + full_text + "']")
            full.click()
            time.sleep(2)
            flag = 0
            
        def step_8():
            firstinput_text = "markFrom"
            firstinput= self.driver.find_element("name", firstinput_text)
            firstinput.send_keys(Keys.RETURN)
            time.sleep(1)
            
        
        steps = [step_1, step_2, step_3, step_4, step_5, step_6, step_7, step_8]
        try:
            self.step_executor(steps)
        except:
            self.driver.refresh()
    
        logging.info(f'Downloaded from {currentfrom} to {currentto}')
        
    def start_wos(self):
        
        driver = self.driver
        webpath = "path-to-wos"
        driver.get(webpath)
        loginUser()

        #=================================================
        cookiebutton = "onetrust-accept-btn-handler"
        l = driver.find_element("id",cookiebutton)
        l.click()
        searchbox = "search-main-box" 
        search = driver.find_element("name", searchbox)
        searchterms = '"Artificial Intelligence" or "Natural Language Processing" or "Computation and Languages"'
        search.send_keys(searchterms)
        time.sleep(1)
        fieldselector_text = "All Fields"
        fieldselector = driver.find_element("xpath", "//div[.='" + fieldselector_text + "']")
        fieldselector.click()
        topic_text = "Topic"
        topic = driver.find_element("xpath", "//div[.='" + topic_text + "']")
        topic.click()
        time.sleep(1)

        searchbutton_text = "mat-focus-indicator search mat-flat-button mat-button-base mat-primary"
        searchbutton = driver.find_element("xpath", f'//button[@class="{searchbutton_text}"]')
        searchbutton.click()
        time.sleep(5)
    def download_batches(self, max_downloads=80000, batch_size = 1000, tries = 3, start=0):
        for i in range(start, max_downloads, batch_size):
            self.download_file(i, i+batch_size-1, tries=tries)
        res = max_downloads%batch_size
        if res>10:
            self.download_file(max_downloads-res, max_downloads, tries=tries)

#================================
wosdownloader = WOSDownloader()
wosdownloader.start_wos()

```
